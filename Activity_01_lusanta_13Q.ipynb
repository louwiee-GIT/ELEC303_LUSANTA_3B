<h2>Question 2: What is the average body mass of male chinstrap from the year 2007? </h2

```python
male_chinstrap_2007 = penguins[(penguins["species"] == "Chinstrap") & (penguins["sex"] == "male") & (penguins["year"] == 2007)]
average_mass = male_chinstrap_2007["body_mass_g"].mean()
male_chinstrap_2007
```

```python
average_mass
```

<h2>Question 4: What is the difference between the heaviest and lightest penguin body mass in the year 2007?</h2

```python
penguins_2007 = penguins[penguins["year"] == 2007]
max_mass = penguins_2007["body_mass_g"].max()
min_mass = penguins_2007["body_mass_g"].min()
mass_difference_2007 = max_mass - min_mass

print(f"Heaviest penguin in 2007: {max_mass} g")
print(f"Lightest penguin in 2007: {min_mass} g")
print(f"Difference: {mass_difference_2007} g")
```

<h3>Insight: This indicates that the heaviest penguin in 2007 has 6.3kg while the lightest has 2.9 kg. This shows that their difference in body mass is 3.4 kg.</h3>
